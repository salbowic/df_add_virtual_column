# Car Dataset Playground

```python
import pandas as pd
from solution import add_virtual_column
```

### 1. Dataset

```python
df_cars = pd.DataFrame({
    'make': ['Porsche', 'Ferrari', 'Lamborghini', 'McLaren', 'Aston Martin', 'Chevrolet'],
    'model': ['911 Turbo S', 'F8 Tributo', 'Huracan Evo', '720S', 'DBS Superleggera', 'Corvette Z06'],
    'base_price': [230000, 280000, 210000, 305000, 316000, 105000],
    'options_price': [45000, 60000, 35000, 75000, 25000, 15000],
    'dealer_discount': [5000, 0, 10000, 15000, 20000, 2000],
    'horsepower': [640, 710, 630, 710, 715, 670],
    'cylinders': [6, 8, 10, 8, 12, 8],
    'weight_lbs': [3636, 3164, 3135, 3128, 3733, 3434],
})
display(df_cars)
```

### 2. Operation: Calculating Subtotal

```python
df_step1 = add_virtual_column(
    df_cars, 
    "base_price + options_price - dealer_discount", 
    "subtotal"
)
display(df_step1)
```

### 3. Operation: Applying Sales Tax (Float Multiplication)

```python
df_step2 = add_virtual_column(
    df_step1, 
    "subtotal + subtotal * 0.085", 
    "out_the_door_price"
)
display(df_step2[['make', 'model', 'subtotal', 'out_the_door_price']])
```

### 4. Operation: Performance Score

```python
df_step3 = add_virtual_column(
    df_step2, 
    "horsepower * cylinders - weight_lbs * 0.5", 
    "performance_score"
)
display(df_step3[['make', 'model', 'horsepower', 'cylinders', 'weight_lbs', 'performance_score']])
```
